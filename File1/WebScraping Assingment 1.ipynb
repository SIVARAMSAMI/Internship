# Question 3

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup = BeautifulSoup(page.content)
team_data = []
table = soup.find("table", class_="table")
rows = table.find_all("tr")

for i in rows[1:11]:
    
    cells = i.find_all("td")
    x = cells[1].text.strip().replace('\n', ' ')
    y = cells[2].text.strip()
    z = cells[3].text.strip()
    w = cells[4].text.strip()
    team_data.append([x, y, z, w])
    
df = pd.DataFrame(team_data, columns=["Team", "Matches", "Points", "Rating"])
print(df)
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup = BeautifulSoup(page.content)
batsmen_ratings = []
table = soup.find("table", class_="table")
rows = table.find_all("tr")

for i in rows[1:11]:
    cells = i.find_all("td")
    bowler = cells[1].text.strip()
    team = cells[2].text.strip()
    rating = cells[3].text.strip()
    batsmen_ratings.append([bowler,team,rating])
    
df = pd.DataFrame(batsmen_ratings, columns =["Batsmen", "Team", "Rating"])
print(df) 
```

```python
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup = BeautifulSoup(page.content)
heading = soup.find_all('a', class_="sc-5smygv-0 fIXTHm")
Title=[]
for i in heading:
    Title.append(i.text)

Title
```

```python
import regex as re
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup = BeautifulSoup(page.content)
Related_Links = []
for link in soup.find_all('a', attrs={'href': re.compile("^https://www.sciencedirect.com/science/article/pii/")}): 
    
    Related_Links.append(link.get('href'))   
print(Related_Links)

Dates = soup.find_all('span', class_="sc-1thf9ly-2 dvggWt")
Published_Date =[]
for i in Dates:
    Published_Date.append(i.text)

print(Published_Date)


Writer = soup.find_all('span', class_="sc-1w3fpd7-0 dnCnAO")
Author =[]
for i in Writer:
    Author.append(i.text)

print(Author)
```

```python
df = pd.DataFrame({'Paper Title':Title, 'Authors':Author, 'Published Date':Published_Date, 'Paper URL':Related_Links })
df
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.content)
names=[]
for i in soup.find_all('a', class_="restnt-name ellipsis"):
    names.append(i.text)

loc=[]
for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    loc.append(i.text)


rating=[]
for i in soup.find_all('div', class_="restnt-rating rating-4"):
    rating.append(i.text)
    

images=[]
for i in soup.find_all('img', class_="no-img"):
    images.append(i['data-src'])
```

```python
cus=[]
for i in soup.find_all('span', class_="double-line-ellipsis"):
    cus.append(i.text)
    
cus
```

```python
import regex as re
data = '₹ 2,000 for 2 (approx) | Chinese, North Indian'
pattern = r'\₹\s\d\,[0-9]*\s[a-z]*\s\d\s\([a-z]*\)\s\|'
cusine= re.sub(pattern,'', data)
cusine
```

```python
cus
```

```python
cusine=[]
for i in cus:
    pattern = r'\₹\s\d\,[0-9]*\s[a-z]*\s\d\s\([a-z]*\)\s\|'
    result= re.sub(pattern,'', i)
    cusine.append(result)
    
```

```python
cusine
```

```python
df = pd.DataFrame({'Restaurant name': names, 'Cuisine': cusine, 'Location': loc,'Ratings': rating, 'Image URL': images})
df
```
