```python
driver = webdriver.Chrome()
#To load url
driver.get("https://www.shine.com/new/job-search")
#job_title = driver.find_element(By.XPATH,"/html/body/div/header[2]/div[2]/div[1]/input")
#job_title.click()
title1 = driver.find_element(By.XPATH,"//input[@id='id_q']")
title1.send_keys("Data Analyst")
location = driver.find_element(By.XPATH,"//input[@id='id_loc']")
location.send_keys("Bangalore")
#time.sleep(5)
search = driver.find_element(By.XPATH,"/html/body/div/div[1]/div/div/div[1]/div[1]/div/div[2]/div/div/form/div/div[2]/div/button")
search.click()

job_title=[]
job_location=[]
company_name=[]
experience=[]

title_ele=driver.find_elements(By.XPATH,"//h2[@itemprop='name']/a")
for i in title_ele[0:10]:
    title=i.text
    job_title.append(title)

print(job_title)

location_ele=driver.find_elements(By.XPATH,"//div[@class='jobCard_jobCard_lists__fdnsc']/div[2]")
for i in location_ele[0:10]:
    location=i.text.replace('\n+14','')
    location1=location.replace('\n+6','')
    location2=location1.replace('\n+9','')
    location3=location2.replace('\n+4','')
    job_location.append(location3)
    
print(job_location)

company_ele=driver.find_elements(By.XPATH,"//div[@class='jobCard_jobCard_cName__mYnow']")
for i in company_ele[0:10]:
    name=i.text
    company_name.append(name)

print(company_name)

experience_ele=driver.find_elements(By.XPATH,"//div[@class=' jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t']")
for i in experience_ele[0:10]:
    exp=i.text
    experience.append(exp)

print(experience)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience))
```

```python
df = pd.DataFrame({'Title': job_title, 'Company_Name': company_name, 'Location': job_location, 'Experience Required': experience})
df
```
