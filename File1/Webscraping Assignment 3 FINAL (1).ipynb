```python
urls=[]

for page in range(0,3):
    url_ele=driver.find_elements(By.XPATH,"//div[@class='a-section a-spacing-base']/div[2]/div/h2/a")
    for i in url_ele:
        x=i.get_attribute('href')
        urls.append(x)
    if(page<2):
        time.sleep(2)
        next_btn=driver.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        next_btn.click()
        time.sleep(2)
```

```python
len(urls)
```

```python
brand=[]
prod_name=[]
exp_delivery=[]
price=[]
availability=[]
exchange=[]

for i in urls:
    driver = webdriver.Chrome()
    driver.get(i)
    time.sleep(9)
    
    try:
        brand_ele=driver.find_element(By.XPATH,"(//span[@class='a-size-base po-break-word'])[1]")
        brand.append(brand_ele.text)
    except NoSuchElementException:
        brand.append('-')
    
    try:
        prod_name_ele=driver.find_element(By.XPATH,"//span[@id='productTitle']")
        prod_name.append(prod_name_ele.text)
    except NoSuchElementException:
        prod_name.append('-')
    
    try:
        exp_delivery_ele=driver.find_element(By.XPATH,"(//div[@class='a-spacing-base']/span/span)[1]")
        exp_delivery.append(exp_delivery_ele.text)
    except NoSuchElementException:
        exp_delivery.append('-')
    
    try:
        price_ele=driver.find_element(By.XPATH,"(//span[@class='a-offscreen'])[1]")
        price.append(price_ele.text)
    except NoSuchElementException:
        price.append('-')
    
    try:
        availability_ele=driver.find_element(By.XPATH,"//div[@id='availability']/span")
        availability.append(availability_ele.text)
    except NoSuchElementException:
        availability.append('-')
    
    try:
        exchange_ele=driver.find_element(By.XPATH,"(//div[@class='a-carousel-viewport']/ol)[2]/li[3]/div/span/div[2]/span")
        exchange.append(exchange_ele.text)
    except NoSuchElementException:
        exchange.append('-')
    time.sleep(2)
    
    
```

```python
print(len(brand),len(prod_name),len(exp_delivery),len(price),len(availability),len(exchange),len(urls))
```

```python
driver = webdriver.Chrome()
#To load url
driver.get("https://www.hostelworld.com/")
driver.maximize_window()
#to type in the search bar
search_bar=driver.find_element(By.XPATH,"//input[@class='native-input']")
search_bar.send_keys("London")
#to click serach content
time.sleep(3)
clk_content=driver.find_element(By.XPATH,"//button[@class='item-content']")
clk_content.click()
#to click the button
time.sleep(3)
btn_letsgo=driver.find_element(By.XPATH,"//button[@class='btn-content large-button']")
btn_letsgo.click()

#creating empty list for all attributes
name=[]
distance=[]
ratings=[]
total_rev=[]
overall_rev=[]
privates_price=[]
dorms_price=[]
hostel_url=[]
facilities=[]
prop_desc=[]

time.sleep(10)
name_ele=driver.find_elements(By.XPATH,"//div[@class='property-name']/span")
for i in name_ele:
    x=i.text
    name.append(x)

distance_ele=driver.find_elements(By.XPATH,"//div[@class='property-distance']/span[2]")
for i in distance_ele:
    x=i.text.replace('-','')
    distance.append(x)

ratings_ele=driver.find_elements(By.XPATH,"//div[@class='rating medium theme-null']/div[1]/span[1]")
for i in ratings_ele:
    x=i.text
    ratings.append(x)

total_rev_ele=driver.find_elements(By.XPATH,"//div[@class='rating medium theme-null']/div[2]/span")
for i in total_rev_ele:
    x=i.text
    total_rev.append(x)

overall_rev_ele=driver.find_elements(By.XPATH,"//div[@class='rating medium theme-null']/div[1]/span[2]")
for i in overall_rev_ele:
    x=i.text
    overall_rev.append(x)

privates_price_ele=driver.find_elements(By.XPATH,"//div[@class='property-accommodation-prices']/div[1]")
for i in privates_price_ele:
    x=i.text.replace('\n',' ')
    x1=x.replace('-36%','')
    x2=x1.replace('-10%','')
    privates_price.append(x2)

dorms_price_ele=driver.find_elements(By.XPATH,"//div[@class='property-accommodation-prices']/div[2]")
for i in dorms_price_ele:
    x=i.text.replace('\n',' ')
    x1=x.replace('-36%','')
    x2=x1.replace('-10%','')
    x3=x2.replace('-5%','')
    dorms_price.append(x3)

hostel_url_ele=driver.find_elements(By.XPATH,"//div[@class='property-list']/div/div/div[4]/div/a")
for i in hostel_url_ele:
    x=i.get_attribute('href')
    hostel_url.append(x)

print(name)
print(distance)
print(ratings)
print(total_rev)
print(overall_rev)
print(privates_price)
print(dorms_price)
print(hostel_url)
```

```python
prop_desc=[]
faclities=[]

for i in hostel_url:
    driver = webdriver.Chrome()
    driver.get(i)
    time.sleep(5)
    try:
        desc_ele=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[2]/div[3]/div/div[2]")
        prop_desc.append(desc_ele.text)
    except NoSuchElementException:
        prop_desc.append('Not present')
    time.sleep(5)
    try:
        faclity_ele=driver.find_element(By.XPATH,"//ul[@class='facility-sections']")
        faclities.append(faclity_ele.text.replace('\n',''))
    except NoSuchElementException:
        faclities.append('Not present')
    
    #driver.quit()

print(prop_desc)   
print(faclities)    
```
